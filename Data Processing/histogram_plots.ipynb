```python
# import the necessary packages
import seaborn as sns
import pandas as pd
```

```python
# the table needs to have these columns: "id" (node id) and "number of links"
# use the relative path to the file name here. If it's in the same directory as the code, just list  the file.
pa = pd.read_csv("test_network.csv")
```

```python
# just checking data
print(pa.head())
```

```python
# this is the plot with bin of size 1
sns.histplot(data=pa, x="number of links").set(title='Pref. Att. bin 1')
```

```python
# we can adjust the bin width, and cut off unnecessary outliers.
sns.histplot(data=pa, x="number of links", binwidth=10, binrange=(0,10000)).set(title='Small. W. bin 10',  xlim=(1,1000))
```

```python
# instead of bins, we can plot a curve for the histogram peaks. 
# This plot also adds log scale to both Y and Y, and cuts off the outliers on X.
sns.histplot(data=pa, x="number of links", element="poly", log_scale=(True, True)).set(title='Small. W. bin 1',  xlim=(1,50000))
```

```python
# This is an example where we can plot multiple data types on one histogram.
sns.histplot(data=pa, x="number of links", binwidth=10, binrange=(0,220), hue="est", multiple="dodge", palette="tab10").set(title='Small W. bin 10')
```
